```python
user = int(input("Enter an angle"))
if user < 90:
    print("Acute")
elif user == 90:
    print("Right")
elif user in range(90,180+1):
    print("Obtuse")
elif user == 180:
    print("Straight")
elif user > 180 and user < 360:
    print("Reflex")
else:
    print("Invalid")
    
```
